## 컬럼을 하나씩 제거하면서 성능확인
- StratifiedKFold cross-validation에서 auc auc_threshold 보다 작아지면 feature 제거를 멈춤

```python
def find_lowest_importance_feature(model, data, features, label):
    X = data[features]
    y = data[label]
    
    model.fit(X, y)
    
    importance_sort_df = pd.DataFrame({
        'feature':X.columns,
        'importance':model.feature_importances_
    }).sort_values(by='importance', ascending=False)
    
    feature = importance_sort_df.iloc[-1, :]['feature']
    importance = importance_sort_df.iloc[-1, :]['importance']
    
    print(f"lowest importance feature: {feature}")
    print(f"importance: {importance}")
    
    return feature
```

```python
def check_auc_per_importance(model_dict, nsplit, data, features, label, threshold, auc_threshold, max_num_feature, min_num_feature):
    n_iter = 1
    while True:
        print(f"{n_iter}번째 iteration")
        print('*'*120)
        kfold_average_metrics_dict, fprs_tprs_dict = stratified_kfold(model_dict=model_dict, 
                                                                 nsplit=nsplit,
                                                                 data=data,
                                                                 features=features, label=label, 
                                                                 threshold = threshold)
    
        auc = kfold_average_metrics_dict['AUC'][0]

        if len(features) < max_num_feature:
            if auc < auc_threshold:
                return before_data
        
        before_data = data.copy()
        
        model = list(model_dict.values())[0]
        lowest_importance_feature = find_lowest_importance_feature(model=model, data=data, features=features, label=label)
        data = data.drop(lowest_importance_feature, axis=1)
        features = data.drop(label, axis=1).columns.tolist()
        
        print(f"남은 feature 개수: {len(features)}")
        if len(features) < min_num_feature:
            return before_data
        
        print('*'*120)
        print()
        
        n_iter += 1
```

```python
threshold = (df[label] == 1).mean() # 실제 데이터에서 1이 차지하는 비율을 threshold로 사용

model_dict = {'LGBM':LGBMClassifier(random_state=0)}

reduced_df = check_auc_per_importance(model_dict=model_dict, 
                                     nsplit=10,
                                     data=df,
                                     features=features, label=label,
                                     threshold=threshold,
                                     auc_threshold=0.785,
                                      max_num_feature=20,
                                      min_num_feature=5)
```

## 남은 feature들의 importance

```python
model = LGBMClassifier(random_state=0
                                    #원하는 파라미터 튜닝을 시도하면 됨
                                   #ex. n_estimators = 100,
                                   #      learning_rate = 0.0001  # parmameter 튜닝 공간
                                                                          )


reduced_X = reduced_df[reduced_features]
reduced_y = reduced_df[label]
X_train, X_test, y_train, y_test = train_test_split(reduced_X, reduced_y, test_size=0.3, stratify=y, random_state=0, shuffle=True)

model.fit(X_train, y_train)
y_prediction = (model.predict_proba(X_test)[:, 1] > y.mean()).astype(int)
y_prediction_proba = model.predict_proba(X_test)[:, 1]
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

feature_importance(model, axes[0],reduced_X.columns.tolist(), color=cm.get_cmap('tab10')(0))
roc_graph(axes[1], y_test, y_prediction_proba)
plt.show()
##fig를 artifect에 저장하는 방법을 고도화 할 예정
```

## MLflow에 로깅

```python
metric_dict= binary_classification_metric(y_test, y_prediction, y_prediction_proba)  ### metric 정의
reduced_features = tag_dict = {f"{idx+1}" :columns for idx, columns in enumerate(reduced_X.columns)} ## reduced_X 자리에 모델에 입력된 X값을 넣어주면 됨

print(f"model: {model}\n")
print(f"model_params_dict: {model.get_params()}\n")  ## 모델 안에 있는 파라미터를 가져옴
print(f"metric_dict: {metric_dict}\n")
print(f"tags_dict: {reduced_features}")
```
